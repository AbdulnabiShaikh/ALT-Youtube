```python
import numpy as mp
import pandas as pd
import random
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
from sklearn.metrics.pairwise import cosine_similarity
```

```python
videos = pd.read_csv('YoutubeVideoDataset.csv')
videos.head(10)
```

```python
vectors = cv.fit_transform(videos['Category']).toarray()
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted_list = sorted(list(enumerate(similarity[0])),key=lambda x:x[1])[0:101]
```

```python
def recommend(video):
  video_index = videos[videos['Category'] == video].index[0]
  distances = similarity[video_index]
  videos_list = sorted(list(enumerate(distances)),key=lambda x:x[1])[0:101]
  #videos_list = random_list

  for i in videos_list:
    print("Title : {}, Category : {}".format(videos.iloc[i[0]].Title, videos.iloc[i[0]].Category))
```

```python
recommend('Food')
```
